```python
string = tf.Variable('This is a string', tf.string)
number = tf.Variable(234, tf.int16)
floating = tf.Variable(3.456, tf.float64)
```

```python
rank1_tensor = tf.Variable(['Test', 'Ok', 'Jobin'], tf.string)
rank2_tensor = tf.Variable([['Test', 'ok'], ['test', 'yes']], tf.string)

print(tf.rank(rank1_tensor))
print(tf.rank(rank2_tensor))
```

```python
tensor1 = tf.ones([1,2,3])
tensor2 = tf.reshape(tensor1, [2, 3, 1])
tensor3 = tf.reshape(tensor2, [3, -1]) 
# -1 tells the tensor to calcuate the size of the dimension in that place this will reshape the tensor to [3, 2]

print(tensor1)
print(tensor2)
print(tensor3)
```
